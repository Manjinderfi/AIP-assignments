# Primer on Lines and Parabolas

A straight line in cartesian coordinate system $\left(x , y\right)$ is represented as

$y = a+bx$,

where $a$ and $b$ (called coefficients) have some concrete values. Let us consider, for example, the line $y = 1 + 2 x$. This means that for any value of $x$ we can compute the corresponding value for $y$ according to this rule. As a result, we get a table of $\left(x , y\right)$ values such as


<div style="margin-left: auto;
            margin-right: auto;
            width: 30%">

| x    |     y     |(x,y) |
|------|:-----------:|------| 
| 0    | 1+2*0 = 1 | (0,1)|
| 1    | 1+2*1 = 3 | (1,3)| 
| 2    | 1+2*2 = 5 | (2,5)| 
| 3    | 1+2*3 = 7 | (3,7)|
| 4    | 1+2*4 = 9 | (4,9)| 
    
</div>


These points can be plotted in a coordinate system to reveal the graph of the line. It should be noted here that one can use any value for x, positive or negative, and get the value for y so there are infinitely many points $\left(x , y\right)$ that make up the line's graph.

Let's plot this line in python using numpy:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3,4])
y = 1 + 2*x
#y = np.array([1,3,5,7,9])
plt.scatter(x,y)
plt.plot(x,y) 
plt.grid(True)
#plt.show()
```

Let's draw the plot for different values of $a$:

```python
y1 = 1 + 2*x
y2 = 2 + 2*x
y3 = 3 + 2*x
#y = np.array([1,3,5,7,9])
plt.scatter(x,y)
plt.plot(x,y1, label=r'$y = 2x + 1$') # r for backslash character to be used in Latex
plt.plot(x,y2, label=r'$y = 2x + 2$') 
plt.plot(x,y3, label=r'$y = 2x + 3$') 
plt.legend()
plt.grid(True)
```

We can see that the lines have the same slope (steepness). But they intersect the $y$-axis at different points.

Now let's draw the plot for different values of $b$:

```python
y1 = 1 + 2*x
y2 = 1 + 3*x
y3 = 1 + 4*x
#y = np.array([1,3,5,7,9])
plt.scatter(x,y)
plt.plot(x,y1, label='y = 2x + 1') 
plt.plot(x,y2, label='y = 3x + 1') 
plt.plot(x,y3, label='y = 4x + 1') 
plt.legend()
plt.grid(True)
```

Note that now The $y$-intercept $a$ remains the same at 1, but the lines have different slopes.

A line is also a polynomial of degree 1 since it contains only terms involving first powers of $x$. It is also called the linear function because its graph is a straight line. Polynomial of degree 2 contains second powers of $x$ (squares of $x$) and it is also called the quadratic polynomial. Cubic polynomial is of degree 3 and it contains cubes of $x$ i.e. terms such as $x^3$.

Next, consider parabolas given by the quadratic polynomial (of degree 2) as

$y = ax^2+bx+c$.

Also here we talk about $y$-intercept, which is now the coefficient $c$. Parabolas with $a>0$ open upward while $a<0$ means that the parabola open downward.
